## 1. Summary

The primary objective of the research is to analyse and improve the churn rate of the customers at OC&Gym. In order to accomplish this objective, an exploratory data analysis (EDA) is conducted to uncover key trends and actionable insights. This procedure includes computing descriptive statistics, feature engineering, and visualizing datasets. Additionally, cohort analysis was conducted using Python. The data was segmented based on various features to analyze customer behaviors of different segments and to observe patterns in the data that could be utilised for making recommendations to the customers. As expected, the result indicates that the churn rate has increased in response to OC&Gym's price hike across all subscription plans. The majority of churned customers were composed of the standard plan subscription members, which contradicts OC&Gym's initial hypothesis of competitor-driven churn. The reason is because if the hypothesis proves to be true, the majority of churned customers would most likely composed of flexible customers whose subscription price is now almost indifferent to the competitor's standard plan, creating a compelling reason for these flexible customers who have little to no switching costs to switch to a more premium brand. As a result, the research has concluded that the main potential root cause of churn could be attributed to OC&Gym's three-month notification rule for standard customers that requires them to provide three-month notice before leaving. In sum, the policy effectively forces these customers to stay, prohibiting them from unsubscribing, and by extension, having to pay for this newly increased price for three months before they can finally unsubscribe. Therefore, the recommendation to OC&Gym is to amend this policy as it is an unjust rule that punishes loyal standard customers and making it appear as the company is trying to take advantage of its customers, which could potentially impact the its brand image and profitability. Specifically, OC&Gym can keep the three-month notification rule. However, in the event of a price increase, the customers will not have to pay for this new price, but instead continue to pay an old rate, prior to the price hike, until the three-month period is over.

## Churn of Members

### Churn Rate Calculations

Churn rate is the number of customers who terminate their relationship with a business in a given time period. It has been calculated by finding the number of customers who left at the end of a time period and dividing it by the number of customers who were present at the beginning of the time period.[1]

##### Example


For the month of December in 2015, Dec-2015,

###### Total Members at the beginning of the Time Period: 

Total Member(Nov-2015) = Total Member(Oct-2015) + Join Member(Nov-2015) – End Member(Nov-2015)

                       = 37591 + 8929 - 8399 
                       = 38121

###### Total Members who leave at the end of the time period: 

Members who end membership in December 2015 = 18165

###### Churn Rate Percentage

Churn rate = End Member(Dec-2015) / Total Member (Nov-2015)

= 18165/38121 = 47.65%

## 5. Discussion


### How could OC&Gym improve their pricing strategy? 

The only way to improve pricing strategy in the long run is to simply assign a price, get feedback from customers and look at how it impacts churn rate, adjust accordingly, and repeat.
 
OC&Gym should slowly raise price in small increments and observe how a combination of metrics (churn rate, net income) changes in response to price change. Simply repeat the process until an inflection price point is discovered, a price point where churn rate increases significantly compared to historical change (and profitability drops).
 
Additionally, they should consider increasing pricing tier (more than just "standard" and "flexible"). This should be offered to affluence group that are price sensitive.

Also, to increase the user experience, the gym should not be fully-packed to the point that it ruins user experience. The analysis shows that there are 3 peak times (morning, after lunch, after work). OC&Gym should try to spread out customers to off-peak times, perhaps by offering discounts or a special package.

Approach (lowering current price and make up for it, in the form of increased members):
A.	Find a historical price point where churn rate was low
B.	Use that price point 

### What happened when the price was increased? 


The churn rate for flexible plan mildly rises while that of standard plan saw a more drastic increase.

### What does this suggest as possible options going forward? 

Focus on standard plan customers. 

Options:
- Loyalty programs
- Amending 3-month lock-up policy - currently standard plan members are forced to stay for 3 months after the price increase. The new policy could 1) decrease lock-up period, 2) members pay the old price during the lock-up.

### In addition to price, what other factors influence whether a member is going to leave? 

-	Lack of loyalty programs - No loyalty programs make it so that customers are not loyal to our brand and is willing to switch anytime they think they found a better service. Customers may also be attracted to other gyms with loyalty programs.
Here, we spend more resources on retaining customers, rather than acquiring new ones by providing rewards to returning customers or start a loyalty program that provides benefits to loyal customers.
-	Unsatisfied with customer service / support - create a review system for customer support. Get rid of bad employees and hire good ones.
-	Customers think our competitor is better - observe what our competitor does better, and try to match or beat it.
-	Location (inconvenient) - we should consider reallocating branches.
-	No longer see the value in our service - we may need to do a better job in maintaining the quality of our equipments + hire better trainers and customer support.

### Key KPIs 

- Monthly churn rate

- Net promotor score: Overall satisfaction, availability of equipment, convenience, utility, etc.

- Peak time / off-peak time demand ratio

- Number of visiting / customer

```python
# group members by TimeofVisit
def visit_time(tv):
    if tv >= datetime.time(6,0) and tv <= datetime.time(11,59):
        return 'morning'
    elif tv >= datetime.time(12,0) and tv <= datetime.time(17,59):
        return 'afternoon'
    elif tv >= datetime.time(18,0) and tv <= datetime.time(20,59):
        return 'evening'
    elif tv>= datetime.time(21,0) and tv <= datetime.time(23,59):
        return 'night'
    elif tv>= datetime.time(0,0) and tv <= datetime.time(5,59):
        return 'latenight'
```

```python
# merge visitationdataPART1 and visitationdataPART1 into vad_merge
vad_merge = pd.concat([vad_1, vad_2])

# set index of vad_merge as CustomerID
vad_merge = vad_merge.set_index('CustomerID')

# change format of TimeofVisit into Year-Month-Day format
vad_merge['TimeofVisit'] = [datetime.time(int(dt.split(':')[0]),int(dt.split(':')[1])) for dt in vad_merge['TimeofVisit']]

# group TimeofVisit
vad_merge['GroupofVisitTime'] = [visit_time(tv) for tv in vad_merge['TimeofVisit']]    

# get dummy variables for GroupofVisitTime
vad_merge = pd.get_dummies(vad_merge, columns=['GroupofVisitTime'])

# replace True by 1 and False by 0 
vad_merge['is_peak'] = [1 if x else 0 for x in vad_merge['peakvisits']]

# generate visitactiondata 
# vad_merge.to_csv(r'C:\Users\surface pro4\Downloads\AIB\A1\visitactiondata.csv')
```
